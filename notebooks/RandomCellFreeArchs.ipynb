```python
def check_uniquness(archs):
    if archs is None:
        return False
    for i in range(1, archs.shape[0]):
        for j in range(0, i):
            if (archs[i] == archs[j]).all():
                print(i, j)
                return False
    return True
```

```python
import numpy as np

archs = None
while not check_uniquness(archs):
    archs = np.random.randint(4, size=(2000, 20, 14))
np.save("/content/drive/MyDrive/SC-Research/RobNets/outputs/architectures.npy", archs)
```

```python
def binary(num, width=2):
    b_num = format(num, 'b')
    if len(b_num) > width:
        raise Exception("Input number is out of range!")
    else:
        return b_num.zfill(width)
```

```python
binaryVec = np.vectorize(binary)
bin_archs = binaryVec(archs)
np.save("/content/drive/MyDrive/SC-Research/RobNets/outputs/binary_architectures.npy", bin_archs)
```
